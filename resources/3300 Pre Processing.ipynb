```python
final_df = pd.DataFrame(columns=['Date','Month','Year','Genre','Hours_watched','Hours_streamed','Peak_viewers','Peak_channels','Streamers','Avg_viewers','Avg_channels','Avg_viewer_ratio','freq'])
```

```python
for index, row in data1.iterrows():
    
    date = "" + str(row['Month']) + "-" + str(row['Year'])
    genres = str(row['Genre']).split(';')
    for genre in genres:
        if(genre == ""): continue
        existed = False
        for index2, row2 in final_df.iterrows():
            if(row2['Genre'] == genre and date == row2['Date']):
                existed = True
                final_df.at[index2,'Hours_watched'] += row['Hours_watched']
                final_df.at[index2,'Hours_streamed'] += row['Hours_Streamed']
                final_df.at[index2,'Peak_viewers'] += row['Peak_viewers']
                final_df.at[index2,'Peak_channels'] += row['Peak_channels']
                final_df.at[index2,'Streamers'] += row['Streamers']
                final_df.at[index2,'Avg_viewers'] += row['Avg_viewers']
                final_df.at[index2,'Avg_channels'] += row['Avg_channels']
                final_df.at[index2,'Avg_viewer_ratio'] += row['Avg_viewer_ratio']
                final_df.at[index2,'freq'] += 1
                break
        if existed:
            freq = final_df.at[index2,'freq']
            final_df.at[index2,'Hours_watched'] = final_df.at[index2,'Hours_watched'] / freq
            final_df.at[index2,'Hours_streamed'] = final_df.at[index2,'Hours_streamed'] / freq
            final_df.at[index2,'Peak_viewers'] = final_df.at[index2,'Peak_viewers'] / freq
            final_df.at[index2,'Peak_channels'] = final_df.at[index2,'Peak_channels'] / freq
            final_df.at[index2,'Streamers'] = final_df.at[index2,'Streamers'] / freq
            final_df.at[index2,'Avg_viewers'] = final_df.at[index2,'Avg_viewers'] / freq
            final_df.at[index2,'Avg_channels'] = final_df.at[index2,'Avg_channels'] / freq
            final_df.at[index2,'Avg_viewer_ratio'] = final_df.at[index2,'Avg_viewer_ratio'] / freq
            break
        
        final_df.loc[len(final_df.index)] = [date,row['Month'],row['Year'],genre,row['Hours_watched'],row['Hours_Streamed'],row['Peak_viewers'],row['Peak_channels'],row['Streamers'],row['Avg_viewers'],row['Avg_channels'],row['Avg_viewer_ratio'],1]
```
